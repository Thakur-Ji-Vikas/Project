```python
data.columns = data.columns.str.upper().str.replace('_','') 
#str.replace('_','') is for replace '_' with " " and ' ' with '_'
# change into upper case
```

```python
data.head()
```

```python
#showing the legendary pokemons
data[data['LEGENDARY']==True].head(5)
```

```python
# chane and set the index to the name attribute
data = data.set_index('NAME')
```

```python
data
```

```python
data.index = data.index.str.replace(".*(?=Mega)","")
```

```python
data.head(10)
```

```python
data
```

```python
# Drop the column with axis = 1 ; axis = 0 is for rows
data = data.drop(['#'],axis=1)
```
